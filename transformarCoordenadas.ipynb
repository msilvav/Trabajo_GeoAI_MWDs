```python
import pandas as pd
from pyproj import Proj, Transformer

# Carga de datos
excel_file = 'ruta/a/tu/archivo.xlsx'
sheet_name = 'nombre_de_hoja'
df = pd.read_excel(excel_file, sheet_name=sheet_name, usecols=['NORTE', 'ESTE'])

# Configuración del proyector
zone = 19
is_southern_hemisphere = True
utm_proj = Proj(proj='utm', zone=zone, south=is_southern_hemisphere, ellps='WGS84')
latlon_proj = Proj(proj='latlong', ellps='WGS84')

# Crear un transformador
transformer = Transformer.from_proj(utm_proj, latlon_proj)

# Función para aplicar la transformación
def convert_utm_to_latlon(row):
    lon, lat = transformer.transform(row['ESTE'], row['NORTE'])
    return pd.Series([lat, lon])

# Aplicar la función a cada fila
df[['latitud', 'longitud']] = df.apply(convert_utm_to_latlon, axis=1)

# Guardar el DataFrame modificado de vuelta al Excel
df.to_excel('ruta/a/tu/archivo_modificado.xlsx', index=False)
```
